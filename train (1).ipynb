```python
# Simple Linear Regression Project
#using Scikit-learn to predict `primmer` based on `age`.
```

```python
## Import Required Libraries
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
from sklearn import linear_model
```

```python
## Load Dataset
```

```python
df = pd.read_csv(r"C:\Users\DELL\Desktop\excel.csv")
```

```python
df
```

```python
## Visualize the Data
```

```python
sns.lmplot(x='age',y='primmer',data=df)
```

```python
## Train Linear Regression Model
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['age']],df['primmer'])
```

```python
reg.predict(pd.DataFrame([[21]], columns=['age']))
```

```python
reg.predict(pd.DataFrame([[50]], columns=['age']))
```

```python
## Model Parameters
```

```python
reg.coef_
```
